```python
import numpy as np
import cv2

matrix = np.array([
    [0, 0, 0, 255, 255],
    [0, 0, 0, 255, 255],
    [0, 0, 0, 255, 255],
    [0, 0, 0, 255, 255],
    [0, 0, 0, 255, 255]
], dtype=np.uint8)

kernelx_prewitt = np.array([[-1, 0, 1],
                            [-1, 0, 1],
                            [-1, 0, 1]])


result_matrix = cv2.filter2D(matrix, cv2.CV_64F, kernelx_prewitt)

print("Original Matrix (Image):")
print(matrix)
print("\nResult Matrix after Prewitt X Filter:")
print(result_matrix)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('gray.jpg', cv2.IMREAD_GRAYSCALE)
kernelx_prewitt = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
kernely_prewitt = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

img_prewittx = cv2.filter2D(img, cv2.CV_64F, kernelx_prewitt)
img_prewitty = cv2.filter2D(img, cv2.CV_64F, kernely_prewitt)

abs_prewittx = cv2.convertScaleAbs(img_prewittx)
abs_prewitty = cv2.convertScaleAbs(img_prewitty)

prewitt_combined = cv2.addWeighted(abs_prewittx, 0.5, abs_prewitty, 0.5, 0)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1) 
plt.imshow(img, cmap='gray') 
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(abs_prewittx, cmap='gray')
plt.title('Prewitt - X (Vertical Edges)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(abs_prewitty, cmap='gray')
plt.title('Prewitt - Y (Horizontal Edges)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(prewitt_combined, cmap='gray')
plt.title('Prewitt - Combined')
plt.axis('off')

plt.tight_layout()
plt.show()
```
