```python
#Task_1
import numpy as np
user_input = input("Enter the array elements separated by spaces: ")
try:
    numbers_list = [int(x) for x in user_input.split()]
    original_array = np.array(numbers_list)
    print("\nOriginal array:")
    print(original_array)
    reverse_array = original_array[::-1]
    print("\nReverse array:")
    print(reverse_array)

except ValueError:
    print("Error: Please enter only numbers separated by spaces.")
```

```python
#Task_2
import numpy as np

try:
    
    input1 = input("Enter the first array elements (separated by spaces): ")
    array1 = np.array([int(x) for x in input1.split()])

    input2 = input("Enter the second array elements (separated by spaces): ")
    array2 = np.array([int(x) for x in input2.split()])

    common_values = np.intersect1d(array1, array2)

    print("\nArray 1:", array1)
    print("Array 2:", array2)
    print("Common values between the two arrays:", common_values)

except ValueError:
    print("Error: Please enter only numbers separated by spaces.")
```

```python
#Task_3
import numpy as np

try:
    user_input = input("Enter the array elements (separated by spaces): ")
    original_array = np.array([int(x) for x in user_input.split()])

    max_index = np.argmax(original_array)

    min_index = np.argmin(original_array)

    print("\nOriginal array:", original_array)
    print("Index of Maximum Value:", max_index)
    print("Index of Minimum Value:", min_index)

except ValueError:
    print("Error: Please enter only numbers separated by spaces.")
```
