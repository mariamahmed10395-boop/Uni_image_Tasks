```python
#Task_1
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Read an RGB image
img = cv2.imread(r'C:\Users\zbook g6\Downloads\image.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 2. Convert image to gray
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 3. Apply solarization to pixel values < 128
# We create a copy to modify
solarized_img = gray_img.copy()
solarized_img[solarized_img < 128] = 255 - solarized_img[solarized_img < 128]

# Display Results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Gray Image")
plt.imshow(gray_img, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Solarized Image (<128)")
plt.imshow(solarized_img, cmap='gray')
plt.show()
```

```python
#Task_2
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two images
x1 = cv2.imread(r'C:\Users\zbook g6\Downloads\image.png', 0)
x2 = cv2.imread(r'C:\Users\zbook g6\Downloads\image2.png', 0)

x2 = cv2.resize(x2, (x1.shape[1], x1.shape[0]))

# Operations
y1 = cv2.add(x1, 2 * x2)
y2 = cv2.add((x1 // 2), cv2.add(x2, 128))
y3 = 255 - x2
y4 = cv2.bitwise_or(x1, x2)

titles = ['y = x1 + 2x2', 'y = (x1/2) + (x2+128)', 'y = 255 - x2', 'y = or(x1, x2)']
images = [y1, y2, y3, y4]

plt.figure(figsize=(12,8))

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
